```python
import numpy as np
import tensorflow as tf
import os
from PIL import Image

```

```python
cur_dir=os.getcwd()
dir=os.path.join(os.getcwd() +'/face/train_split_85/0' )
new_dir=os.path.join(cur_dir +'/face/train_split_85_rgb/0' )
files=os.listdir(dir)
file=['/root/UoE-MLP/baseline/face/train_split_85/0/'+ i for i in files]
def read_image(filename_queue):
    reader = tf.WholeFileReader()
    key,value = reader.read(filename_queue)
    image = tf.image.decode_jpeg(value)
    return image   
def gray_to_RGB(image):
    new_img=tf.image.grayscale_to_rgb(image)
    return new_img
def inputs():
    filenames = file
    filename_queue = tf.train.string_input_producer(filenames)
    read_input = read_image(filename_queue)
    rgb_image = gray_to_RGB(read_input)
    return rgb_image
```

```python
with tf.Graph().as_default():
    image = inputs()
    init = tf.global_variables_initializer()
    sess = tf.Session()
    sess.run(init)
    tf.train.start_queue_runners(sess=sess)
    for i in range(5099):
        img = sess.run(image)
        img = Image.fromarray(img, "RGB")
        img.save(os.path.join(new_dir, str(i)+".jpg"))
```
